The sub-metric I will be looking at is the area of each neighborhood of Pittsburgh. The exact data can be found in LandUsePittNeighborhoods.csv. Visualization of the data will be added below using the "pandas" library in Python. The visualization will first include a bar graph showing the land area in acres of each neighborhood. The following graph will show the same data but organized in descending order to show which neighborhood has the most and least land area. I have chosen land area as my metric because having more space is a trait that usually improves quality of life. A larger land area can also be related to increased resources in that neighborhood. 

```python
import pandas as pd

df = pd.read_csv('LandUsePittNeighborhoods.csv')
ax = df.plot.bar(x='Neighborhood', y='Land Area (acres)', figsize=(20, 6), legend=False)
ax.set_xlabel('Neighborhood', fontsize=8)
ax.set_ylabel('Land Area (acres)')
ax.set_title('Land Area by Neighborhood')
plt.xticks(rotation=90, fontsize=8)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('LandUsePittNeighborhoods.csv')

df = df.sort_values(by='Land Area (acres)', ascending=False)
df.reset_index(drop=True, inplace=True)

ax = df.plot.bar(x='Neighborhood', y='Land Area (acres)', figsize=(20, 6), legend=False)
ax.set_xlabel('Neighborhood', fontsize=8)
ax.set_ylabel('Land Area (acres)')
ax.set_title('Land Area by Neighborhood')
plt.xticks(rotation=90, fontsize=8)
plt.show()
```

After viewing the organized bar chart of land area, the conclusion is that Squirrel Hill South has the largest land area. This means it is the "best" neighborhood because it has the most space for people to live comfortably. Therefore, Mt. Oliver would be the "worst" neighborhood because it is the smallest.
