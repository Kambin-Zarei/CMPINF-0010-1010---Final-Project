My sub-metric that I will specifically be focusing on is Population within each neighborhood in the city of Pittsburgh. The exact data can be found in Population.csv. Visualization of the data will be added below using "pandas" library in Python. The visualization will first include a bar graph that is not ordered to show variability within the neighborhoods and how much their populations can differ. And secondly, another bar graph will be generated that shows the same data in DESCENDING order to see exactly which neighborhoods are "best" and "worst" in the city of Pittsburgh. I have chosen "POPULATION" as my metric because the more people that live in an area, the more friends everyone can have. Also, living in a neighborhood with very few people can also be very lonely and sometimes dangerous.

```python
import pandas as pd

df = pd.read_csv('Population.csv')
ax = df.plot.bar(x='Neighborhood', y='Estimate; Total', figsize=(20, 6), legend=False)
ax.set_xlabel('Neighborhood', fontsize=8)
ax.set_ylabel('Estimate; Total')
ax.set_title('Population Estimate by Neighborhood')
plt.xticks(rotation=90, fontsize=8)
plt.show()

# Explanation: I am reading the data from Population.csv into a Pandas DataFrame. 
# Then I am actually creating the bar graph with the x-axis being the neighborhood names and the
# y-axis being the estimated total population of each neighborhood. 
# I also made the figsize especially big in the x-axis (20, 6). 20 representing the x-axis.
# This is because many of the names of the neighborhoods overlapped, so I had to stretch the graph along the 
# x-axis. Then I am simply setting the x-axis and y-axis labels and font size.
# Then, I am setting the title, font, and size. I am also rotating the entire graph 90 degrees
# so that the x-axis is at the bottom and is easier to read. Finally, I use
# plt.show() to display the graph.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Population.csv')

df = df.sort_values(by='Estimate; Total', ascending=False)
df.reset_index(drop=True, inplace=True)

ax = df.plot.bar(x='Neighborhood', y='Estimate; Total', figsize=(20, 6), legend=False)
ax.set_xlabel('Neighborhood', fontsize=8)
ax.set_ylabel('Estimate; Total')
ax.set_title('Population Estimate by Neighborhood')
plt.xticks(rotation=90, fontsize=8)
plt.show()

# Everything here is the same. Except one difference. Just two more lines of code.
# The first line is to sort the DataFrame by the "Estimate Total" column in 
# descending order. That is why "ascending" attribute is set to False.
# The second line is needed to maintain a continous range so the graph flows nicely.
```

After the creation of the sorted graph in descending order, the conclusion is that Squirrel Hill South has the highest population therefore, is the "Best" neighborhood according to solely my metric. And in that case, Chateau would be the worst neighborhood in the city of Pittsburgh because it is the lowest on the list, and its bar is barely registered on the graph.
